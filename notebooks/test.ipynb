```python
def tree_from_dict(d):
    if not d: return {}
    k = next(iter(d))
    v = d.pop(k)
    if not isinstance(v, list): v = [v,]
    return {f"{k}={'/'.join(v)}" : tree_from_dict(d)}

t1 = Tree.from_dict(tree_from_dict({'class': ['d1', 'd2'],
 'dataset': 'climate-dt',
 'generation': ['1','2','3'],
 'model': 'icon',
 'date': ['20241102','20241103'],
 'resolution': ['high','low'],
 'time': ['0000', '0600', '1200', '1800'],
}))
print(str(t1))
```

```python


def parse_key_value_pairs(text: str):
    result = {}
    text = text.replace("}{", ",")  # Replace segment separators
    text = text.replace("{", "").replace("}","").strip()  # Remove leading/trailing braces

    for segment in text.split(","):
        if "=" not in segment: print(segment)
        key, values = segment.split("=", 1)  # Ensure split only happens at first "="
        values = values.split("/")
        result[key] = values

    return result

def parse_fdb_list(f):
    for line in f.readlines():
        # Handle fdb list normal
        if line.startswith("{"):
            yield parse_key_value_pairs(line)

        # handle fdb list --compact
        if line.startswith("retrieve,") and not line.startswith("retrieve,\n"):
            line = line[9:]
            yield parse_key_value_pairs(line)
        

q = Qube.empty()
with gzip.open("/Users/math/git/rust/qubed/tests/data/fdb_list_compact.gz", mode = "rt") as f:
    for i, datacube in enumerate(parse_fdb_list(f)):
        new_branch = Qube.from_datacube(datacube)
        new_q = q | new_branch
        assert new_q.n_leaves == q.n_leaves + new_branch.n_leaves
        q = new_q
        if i % 10 == 0:
            print(f"Done: {i}, Nodes in tree: {q.n_nodes} Leaves in tree: {q.n_leaves}")

q.n_leaves
```
